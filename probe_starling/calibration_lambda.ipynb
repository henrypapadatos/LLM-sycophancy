```python
import pandas as pd
import ast
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

file = '/nas/ucb/henrypapadatos/Sycophancy/results/Bo32_openchat_openchat_3.5_02_16_2024_00_08_55_feedback_poem_300-800_Starlingnew_probeV21.csv'

df = pd.read_csv(file)
df['RM_score'] = df['RM_score'].apply(ast.literal_eval)
df['RM_score'] = df['RM_score'].apply(lambda x: np.array(x))

df['probe_score'] = df['probe_score'].apply(ast.literal_eval)
df['probe_score'] = df['probe_score'].apply(lambda x: np.array(x))

df.head(1)
```

```python
df['centered_RM_score'] = df['RM_score'].apply(lambda x: x - np.mean(x))
all_centered_scores = []
for i in range(len(df)):
    all_centered_scores.extend(df['centered_RM_score'][i])

sns.histplot(all_centered_scores, kde=True)
#set title
plt.title('Centered RM scores distribution')

all_centered_scores = np.array(all_centered_scores)
rm_std = np.std(all_centered_scores)
```

```python
df['centered_probe_score'] = df['probe_score'].apply(lambda x: x - np.mean(x))
all_centered_probe_scores = []
for i in range(len(df)):
    all_centered_probe_scores.extend(df['centered_probe_score'][i])

sns.histplot(all_centered_probe_scores, kde=True)
#set title
plt.title('Centered probe scores distribution')

all_centered_probe_scores = np.array(all_centered_probe_scores)
probe_std = np.std(all_centered_probe_scores)
```

```python
lambda_ = 0.5
df['scaled_probe_score'] = df['probe_score'].apply(lambda x: x*lambda_)
df['centered_scaled_probe_score'] = df['centered_probe_score'].apply(lambda x: x*lambda_)
all_scaled_probe_scores = []
for i in range(len(df)):
    all_scaled_probe_scores.extend(df['centered_scaled_probe_score'][i])

sns.histplot(all_scaled_probe_scores, kde=True)
#set title 
plt.title('Scaled probe scores distribution')

all_scaled_probe_scores = np.array(all_scaled_probe_scores)
scaled_probe_std = np.std(all_scaled_probe_scores)
```

```python
print('RM std:', rm_std)
print('Probe std:', probe_std)
print('When lambda is', lambda_, 'the std of the centered probe score is', scaled_probe_std)
```
